```python
import openai
import json
import os

def read_descriptions(file_path):
    # 파일을 읽기 모드로 열기
    with open(file_path, 'r', encoding='utf-8') as file:
        # JSON 데이터 로드
        data = json.load(file)
    
    return data

def call_chatgpt(word_content):
    # 환경변수에서 API 키 불러오기
    api_key = 'sk-al8LY2KP9JUls1HdCrdrhA2GofI2OztrBR3w1W6pErT3BlbkFJsLFqgNeJKhTgVR3VIBIHVaZ1qvi5FNsS_cBG7rTGIA'
    openai.api_key = api_key
    
    content = """너는 외환 전문가야. 외화 중에서도 특히 "USD/KRW" 전문가야. 사용자가 단어를 주면 그 단어가 "USD/KRW" 투자자입장에서 "긍정"인지 "부정"인지를 답해줘야 해. 다른 답은 하지말고, "긍정" 아니면 "부정"만 답해야 해."""
    
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": content},
                {"role": "user", "content": word_content}
            ],
            max_tokens=1000
        )

        response_content = response['choices'][0]['message']['content'] if 'message' in response['choices'][0] else response['choices'][0]['text']
        
        return response_content.strip()  # 감정 분석 결과를 반환
    except Exception as e:
        print(f"API 호출 중 에러 발생: {e}")
        return None

def save_results(file_path, data):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=2)
    print("데이터가 성공적으로 저장되었습니다.")

def main():
    file_path_read = '/Users/yang-jin-an/top_20_word_frequency_filtered.json'
    file_path_write = 'word_data_usd.json'
    
    # 기존 JSON 데이터를 읽어옴
    word_data = read_descriptions(file_path_read)
    results = []

    for item in word_data:
        word = item['text']
        value = item['value']
        sentiment = call_chatgpt(word)
        
        if sentiment:
            results.append({
                "text": word,
                "value": value,
                "sentiment": sentiment
            })
    
    # 결과를 저장
    save_results(file_path_write, results)

if __name__ == "__main__":
    main()
```
