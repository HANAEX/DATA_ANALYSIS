```python
import json
import mysql.connector

# JSON 파일 경로
json_file_path = "/Users/yang-jin-an/r_exchange_data_JPY.json"

# MySQL 연결 정보 설정
config = {
    'user': 'yangjinan',
    'password': 'ALsk85200!!',
    'host': 'hanaex.mysql.database.azure.com',
    'database': 'my_database',
    'port': 3306
}

# JSON 파일에서 데이터 읽기
with open(json_file_path, 'r', encoding='utf-8') as file:
    data_list = json.load(file)

# MySQL에 연결 및 데이터 삽입
try:
    connection = mysql.connector.connect(**config)
    cursor = connection.cursor()

    print("연결 성공!")

    # 데이터 삽입 SQL 쿼리 준비
    insert_query = """
    INSERT INTO t_ex_table (
        state,
        period,
        time,
        cash_buy,
        cash_sell,
        remit_send,
        remit_receive,
        foreign_currency_check_sell,
        exchange_rate,
        compared_to_previous,
        exchange_fee_rate,
        usd_conversion_rate
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    # 데이터 삽입
    state_value = "JPY"  # state를 USD로 설정
    for row in data_list:
        period = int(row[0])  # period를 JSON 데이터의 첫 번째 값으로 사용
        time_value = row[1]   # time을 JSON 데이터의 두 번째 값으로 사용
        # 나머지 데이터를 JSON의 나머지 값으로 설정
        data_tuple = (
            state_value,
            period,
            time_value,
            row[2],  # cash_buy
            row[3],  # cash_sell
            row[4],  # remit_send
            row[5],  # remit_receive
            row[6],  # foreign_currency_check_sell
            row[7],  # exchange_rate
            row[8],  # compared_to_previous
            row[9],  # exchange_fee_rate
            row[10]  # usd_conversion_rate
        )
        print(data_tuple)
        cursor.execute(insert_query, data_tuple)
    
    # 변경 사항 커밋
    connection.commit()
    print("데이터 삽입 완료!")

except mysql.connector.Error as err:
    print(f"연결 실패 또는 데이터 삽입 실패: {err}")
finally:
    # 커서와 연결 종료
    if cursor:
        cursor.close()
    if connection:
        connection.close()
    print("연결 종료.")
```
